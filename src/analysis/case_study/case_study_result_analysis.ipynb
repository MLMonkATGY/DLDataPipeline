```python
import pandas as pd
```

```python
caseStudyDf = pd.read_csv("/home/alextay96/Desktop/all_workspace/new_workspace/DLDataPipeline/data/case_study/case_study_8/case_study_result.csv")
```

```python
labelErrorDf = caseStudyDf[caseStudyDf["remarks"] == "label error"]
labelErrorByPartDf = labelErrorDf.groupby(["part"])["CaseID"].count().reset_index().rename(columns={"CaseID" : "count"})
labelErrorByPartDf.sort_values(by="count", ascending=False, inplace=True)
labelErrorByPartDf["ratio"] = labelErrorByPartDf["count"] / len(caseStudyDf["CaseID"].unique())
labelErrorByPartDf
```

```python
predErrorDf = caseStudyDf[caseStudyDf["remarks"] == "error"]
predErrorByPartDf = predErrorDf.groupby(["part", "fp", "fn"])["CaseID"].count().reset_index().rename(columns={"CaseID" : "count"})
predErrorByPartDf.sort_values(by="count", ascending=False, inplace=True)
predErrorByPartDf["ratio"] = predErrorByPartDf["count"] / len(caseStudyDf["CaseID"].unique())
predErrorByPartDf
```

```python
tempDf = caseStudyDf[caseStudyDf["remarks"] == "label error"]
countDf = tempDf.groupby("CaseID")["remarks"].count().reset_index().rename(columns={
    "remarks" : "label_error_count"
})
countDf["after_repair_img"] = countDf["label_error_count"].apply(lambda x : x > 2)
countDf[countDf["after_repair_img"] == True]
```

```python
labelErrorButCorrectDf = caseStudyDf[(caseStudyDf["remarks"] == "label error") & (caseStudyDf["pred"] == caseStudyDf["gt"] )]
labelErrorButCorrectByPartDf = labelErrorButCorrectDf.groupby(["part"])["CaseID"].count().reset_index().rename(columns={"CaseID" : "count"})
labelErrorButCorrectByPartDf.sort_values(by="count", ascending=False, inplace=True)
labelErrorButCorrectByPartDf["ratio"] = labelErrorButCorrectByPartDf["count"] / len(caseStudyDf["CaseID"].unique())
labelErrorButCorrectByPartDf
```

```python
labelErrorButIncorrectDf = caseStudyDf[(caseStudyDf["remarks"] == "label error") & (caseStudyDf["pred"] != caseStudyDf["gt"] )]
labelErrorButIncorrectByPartDf = labelErrorButIncorrectDf.groupby(["part"])["CaseID"].count().reset_index().rename(columns={"CaseID" : "count"})
labelErrorButIncorrectByPartDf.sort_values(by="count", ascending=False, inplace=True)
labelErrorButIncorrectByPartDf["ratio"] = labelErrorButIncorrectByPartDf["count"] / len(caseStudyDf["CaseID"].unique())
labelErrorButIncorrectByPartDf
```
