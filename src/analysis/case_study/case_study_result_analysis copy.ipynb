```python
import pandas as pd
```

```python
caseStudyDf = pd.read_csv("/home/alextay96/Desktop/all_workspace/new_workspace/DLDataPipeline/data/case_study/case_study_8/case_study_result.csv")
```

```python
labelErrorDf = caseStudyDf[caseStudyDf["remarks"] == "label error"]
labelErrorByPartDf = labelErrorDf.groupby(["part"])["CaseID"].count().reset_index().rename(columns={"CaseID" : "count"})
labelErrorByPartDf.sort_values(by="count", ascending=False, inplace=True)
labelErrorByPartDf["ratio"] = labelErrorByPartDf["count"] / len(caseStudyDf["CaseID"].unique())
labelErrorByPartDf
```

```python
predErrorDf = caseStudyDf[caseStudyDf["remarks"] == "error"]
predErrorByPartDf = predErrorDf.groupby(["part", "fp", "fn"])["CaseID"].count().reset_index().rename(columns={"CaseID" : "count"})
predErrorByPartDf.sort_values(by="count", ascending=False, inplace=True)
predErrorByPartDf["ratio"] = predErrorByPartDf["count"] / len(caseStudyDf["CaseID"].unique())
predErrorByPartDf
```

```python
tempDf = caseStudyDf[caseStudyDf["remarks"] == "label error"]
countDf = tempDf.groupby("CaseID")["remarks"].count().reset_index().rename(columns={
    "remarks" : "label_error_count"
})
countDf["after_repair_img"] = countDf["label_error_count"].apply(lambda x : x > 2)
countDf[countDf["after_repair_img"] == True]
```

```python
afterRepairError = 8 / 30
afterRepairError
```

```python
hardsampleError = caseStudyDf[(caseStudyDf["v_pred"] != -1) &(caseStudyDf["remarks"] == "hard sample") & (caseStudyDf["pred"] != caseStudyDf["gt"] )]
hardSampleErrorBypartDf = hardsampleError.groupby(["part"])["CaseID"].count().reset_index().rename(columns={"CaseID" : "count"})
hardSampleErrorBypartDf.sort_values(by="count", ascending=False, inplace=True)
hardSampleErrorBypartDf["hard_sample_error"] = hardSampleErrorBypartDf["count"] / len(caseStudyDf["CaseID"].unique())
hardSampleErrorBypartDf
```

```python
labelErrorButCorrectDf = caseStudyDf[(caseStudyDf["v_pred"] != -1) & (caseStudyDf["remarks"] == "label error") & (caseStudyDf["pred"] == caseStudyDf["gt"] )]
labelErrorButCorrectByPartDf = labelErrorButCorrectDf.groupby(["part"])["CaseID"].count().reset_index().rename(columns={"CaseID" : "count"})
labelErrorButCorrectByPartDf.sort_values(by="count", ascending=False, inplace=True)
labelErrorButCorrectByPartDf["label_error_same_ratio"] = labelErrorButCorrectByPartDf["count"] / len(caseStudyDf["CaseID"].unique())
labelErrorButCorrectByPartDf
```

```python
labelErrorButIncorrectDf = caseStudyDf[(caseStudyDf["v_pred"] != -1) &(caseStudyDf["remarks"] == "label error") & (caseStudyDf["pred"] != caseStudyDf["gt"] )]
labelErrorButIncorrectByPartDf = labelErrorButIncorrectDf.groupby(["part"])["CaseID"].count().reset_index().rename(columns={"CaseID" : "count"})
labelErrorButIncorrectByPartDf.sort_values(by="count", ascending=False, inplace=True)
labelErrorButIncorrectByPartDf["label_error_different_ratio"] = labelErrorButIncorrectByPartDf["count"] / len(caseStudyDf["CaseID"].unique())
labelErrorButIncorrectByPartDf
```

```python
pd.set_option('display.float_format', lambda x: '%.3f' % x)

```

```python
majorPartErrorDf = caseStudyDf[(caseStudyDf["v_pred"] != -1) & (caseStudyDf["pred"] != caseStudyDf["gt"] )]
majorErrorByPartDf = majorPartErrorDf.groupby(["part"])["CaseID"].count().reset_index().rename(columns={"CaseID" : "count"})
majorErrorByPartDf.sort_values(by="count", ascending=False, inplace=True)
majorErrorByPartDf["error_ratio"] = majorErrorByPartDf["count"] / len(caseStudyDf["CaseID"].unique())

tempDf = majorErrorByPartDf.merge(labelErrorButIncorrectByPartDf[["part", "label_error_different_ratio"]], on="part", how="outer")
tempDf = tempDf.merge(labelErrorButCorrectByPartDf[["part", "label_error_same_ratio"]], on="part",  how="outer")
tempDf = tempDf.merge(hardSampleErrorBypartDf[["part", "hard_sample_error"]], on="part",  how="outer")
tempDf.fillna(0, inplace=True)
tempDf["unexplained_error"]= tempDf.apply(lambda x :float(x["error_ratio"] - x["hard_sample_error"] - x["label_error_different_ratio"] + x["label_error_same_ratio"])  , axis=1)
tempDf.sort_values(by="unexplained_error", ascending=False)
```
