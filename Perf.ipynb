```python
import pandas as pd
import plotly.express as px
import numpy as np
```

```python
evalDf = pd.read_csv("/home/alextay96/Desktop/new_workspace/DLDataPipeline/perf_breakdown.csv")
```

```python
meanVal = evalDf["caseAcc"].mean()
medianVal = evalDf["caseAcc"].median()
firstQuarter = np.quantile(evalDf["caseAcc"].values, 0.25)
thirdQuarter = np.quantile(evalDf["caseAcc"].values, 0.75)

fig = px.histogram(evalDf, x="caseAcc")
fig.add_vline(x=meanVal, line_width=2, line_color="red")
fig.add_vline(x=medianVal, line_width=2, line_color="blue")
fig.add_vline(x=firstQuarter, line_width=2, line_color="orange")
fig.add_vline(x=thirdQuarter, line_width=2, line_color="green")

fig.update_layout(
    xaxis = dict(
        tickmode = 'array',
        tickvals = [meanVal, medianVal, firstQuarter, thirdQuarter],
    )
)
fig.show()
```
